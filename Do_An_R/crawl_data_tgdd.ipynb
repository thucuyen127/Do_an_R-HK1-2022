```python
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
import csv
from datetime import date
import random
import requests
```

```python
# open browser

PATH = "chromedriver.exe"
driver = webdriver.Chrome(PATH)
```

```python
#lists of output
item_id = []
item_name = []
price = []
category = []
urls = []
total_ratings = []
brands = []

#lists of output2
name_review = []
cate = []
content = []
brand_of_item = []
rate_score = []
```

```python
def craw_data_dtdd():
    driver.get('https://www.thegioididong.com/dtdd#c=42&o=9&pi=7')
    sleep(5)


    try:
        page = driver.find_element_by_xpath('//*[@id="categoryPage"]/div[3]/div[2]/a')
        page.click()
        sleep(5)
    except:
        pass

    page_source = BeautifulSoup(driver.page_source,  'html.parser')
    items = page_source.find_all('li', class_ = 'item ajaxed __cate_42')
    for item in items:
        tag_a = item.find('a')
        url1 = tag_a.get('href')
        url = 'https://www.thegioididong.com' + url1
        urls.append(url)
        it_id = int(item.get('data-id'))
        item_id.append(it_id)
        prc = (item.get('data-price'))
        prc = int(prc[:-2])
        price.append(prc)
        name = item.find('h3').text
        item_name.append(name.strip())
        category.append('Smartphone')
        try:
            ratings = int(item.find('p', class_ = 'item-rating-total').text)
        except:
            ratings = 0

        total_ratings.append(ratings)
```

```python
def craw_data_chuot():
    driver.get('https://www.thegioididong.com/chuot-may-tinh#c=86&o=8&pi=3')
    sleep(5)



    page_source = BeautifulSoup(driver.page_source,  'html.parser')
    items = page_source.find_all('li', class_ = 'item __cate_86')
    for item in items:
        tag_a = item.find('a')
        url1 = tag_a.get('href')
        url = 'https://www.thegioididong.com' + url1
        urls.append(url)
        it_id = int(item.get('data-id'))
        item_id.append(it_id)
        prc = (item.get('data-price'))
        prc = int(prc[:-2])
        price.append(prc)
        name = item.find('h3').text
        item_name.append(name.strip())
        category.append('Basic mouse')
        try:
            ratings = int(item.find('p', class_ = 'item-rating-total').text)
        except:
            ratings = 0

        total_ratings.append(ratings)
    return len(items)
```

```python
def data_reviewer(url, i, item_id, brand):
    url = url[:-8] + '/danh-gia'
    driver.get(url)
    sleep(4)
    z = 1

    try:
        next_ = driver.find_elements_by_xpath('//div[@class = "pagcomment"]/a')
        j = len(next_)
        h = int(next_[-2].text)
    except:
        h = 0


    while True:
        page_source = BeautifulSoup(driver.page_source,  'html.parser')
        reviews = page_source.find_all('div', class_ = 'comment__item par')
        for review in reviews:
            name2 = review.find('p', class_ = 'txtname').text
            name_review.append(name2)
            cate.append(category[i])
            brand_of_item.append(brand)
            cont = review.find('p', class_ = "cmt-txt").text
            content.append(cont)
            s_rate = review.find_all('i', class_ = 'icon-star')
            _rate = len(s_rate)
            rate_score.append(_rate)
            score = float(driver.find_element_by_xpath('//p[@class = "point"]').text)

        
        if z < h:
            next_[j - 1].click()
            sleep(3)
            next_ = driver.find_elements_by_xpath('//div[@class = "pagcomment"]/a')
            j = len(next_)
            z += 1
        else:
            break
    return score
    
```

```python
craw_data_dtdd()
craw_data_tablet()
craw_data_sac()
craw_data_watch()
m = craw_data_chuot()
```

```python
n = len(item_id)


def brand_i(name, i):
    if i < (n - m):
        find_ = name.find(' ')
        brand = name[0:find_]
        if (brand == 'iPhone') or (brand == 'iPad'):
            brand = 'Apple'
    else:
        name = name.replace('Chuột ', '').replace('Có ', '').replace('Không ', '').replace('Dây ', '').replace('Gaming ', '').replace('Silent ', '').replace('Bluetooth ', '').replace('Rút ', '')
        find_ = name.find(' ')
        brand = name[0:find_]
    return brand
```

```python
i = 0
with open('output.csv', 'w', encoding='UTF8',  newline = '') as file_output:
    headers = ['Item ID', 'Item Category', 'Item Name', 'Brand Name', 'Item URL', 'Price','Rate', 'Total Reviews', 'Retrieved Date']
    writer = csv.DictWriter(file_output, delimiter=',', lineterminator='\n',fieldnames=headers)
    writer.writeheader()
    for i in range(n):
        today = date.today()
        retrived_date = today.strftime("%m/%d/%Y")
        brand = brand_i(item_name[i], i)
        if total_ratings[i] > 0:
            score = data_reviewer(urls[i], i, item_id, brand)
        else:
            score = 0
        if price[i] > 0:
            writer.writerow({headers[0]: item_id[i],headers[1]: category[i], headers[2]: item_name[i], headers[3]: brand, headers[4]: urls[i] , headers[5]: price[i], headers[6]: score, headers[7]: total_ratings[i], headers[8]: retrived_date})
            i += 1
        print('cào lần thứ', i)
```

```python
i = 0
with open('output2.csv', 'w', encoding='UTF8',  newline = '') as file_output:
    headers = ['Name of reviewer', 'Category','rate', 'content', 'brand of Item']
    writer = csv.DictWriter(file_output, delimiter=',', lineterminator='\n',fieldnames=headers)
    writer.writeheader()
    for i in range(len(name_review)):
        writer.writerow({headers[0]: name_review[i], headers[1]: cate[i], headers[2]: rate_score[i], headers[3]: content[i], headers[4]: brand_of_item[i]})
```
